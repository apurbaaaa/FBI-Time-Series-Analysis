# **Project Summary -**

This project aims to analyze crime patterns by exploring the relationship between crime type, neighborhood, and the time of occurrence using statistical techniques. Understanding these relationships is crucial for law enforcement, policymakers, and researchers to develop effective crime prevention strategies and optimize resource allocation. The study follows a structured approach, including data preprocessing, outlier handling, and hypothesis testing to derive meaningful insights from crime data.

Data Preprocessing and Outlier Handling
Before conducting any statistical tests, the dataset underwent preprocessing to ensure data quality and reliability. Handling outliers was a key step in this process since extreme values could distort the analysis. The Interquartile Range (IQR) method was used to detect outliers. This method involves calculating the first quartile (Q1) and the third quartile (Q3) of numerical features, with the interquartile range (IQR) being the difference between these values. Any data point below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR was considered an outlier. Instead of removing these outliers, the project implemented a clipping technique, where values were adjusted to the nearest valid boundary within the acceptable range. This ensured that extreme values did not unduly influence the statistical tests while preserving the overall data distribution.

Hypothesis Testing for Crime Analysis
The core of this project involved conducting hypothesis tests to understand the influence of different factors on crime patterns. Three hypothetical statements were tested using appropriate statistical techniques to determine significant associations in the data.

Hypothetical Statement 1: Relationship Between Neighborhood and Crime Type

To examine whether the type of crime is associated with specific neighborhoods, a Chi-Square Test of Independence was conducted. This test evaluates whether two categorical variables are significantly related. The contingency table of crime type and neighborhood was created, and the Chi-Square statistic and p-value were calculated. The results showed a significant association between crime type and neighborhood, as the p-value was less than 0.05. This indicates that certain crime types are more prevalent in specific neighborhoods, suggesting that crime patterns are influenced by location. Law enforcement agencies can use this insight to allocate resources effectively and implement targeted crime prevention measures in high-risk areas.

Hypothetical Statement 2: Influence of Neighborhood on Crime Occurrence Time

To investigate whether the time at which crimes occur varies across neighborhoods, an ANOVA (Analysis of Variance) test was performed. ANOVA is used to compare means across multiple groups to determine whether significant differences exist. The test results yielded a p-value below 0.05, leading to the rejection of the null hypothesis. This suggests that crime occurrence time significantly differs depending on the neighborhood. This insight is valuable for law enforcement as it helps in identifying high-crime time periods for specific locations, allowing for better strategic planning of patrolling schedules and resource distribution.

Hypothetical Statement 3: Impact of Crime Type on Crime Occurrence Time

Another ANOVA test was conducted to explore whether the occurrence time of crimes is significantly influenced by the type of crime. The statistical test showed a strong correlation between crime type and time of occurrence, with a very low p-value indicating statistical significance. This finding implies that different types of crimes tend to occur at distinct times of the day. For example, certain violent crimes might be more common at night, while theft-related crimes might peak during busy hours. Understanding this pattern helps law enforcement agencies design time-sensitive intervention strategies and optimize their surveillance efforts accordingly.

Conclusion and Implications
The results of this project demonstrate that crime patterns are significantly influenced by location and time. The statistical analyses confirm that crime occurrence varies across neighborhoods, and different types of crimes tend to happen at specific times. These findings highlight the importance of data-driven approaches in crime prevention. By identifying high-crime neighborhoods and peak crime times, law enforcement agencies can optimize patrolling strategies, improve emergency response planning, and allocate resources more efficiently. Additionally, policymakers can use these insights to implement community-based crime prevention programs in areas with high crime rates.

Moreover, this study emphasizes the role of statistical techniques in criminology. The application of Chi-Square and ANOVA tests provided empirical evidence supporting the existence of spatial and temporal crime trends. These methods can be further extended by incorporating machine learning algorithms to predict crime hotspots and analyze additional influencing factors such as weather, economic conditions, and demographic characteristics.

In summary, this project successfully utilized statistical analysis to uncover significant crime patterns. The findings reinforce the need for targeted crime prevention efforts based on data-driven insights. Future work can explore deeper relationships by integrating more complex models and larger datasets, further enhancing crime prediction and prevention strategies.

````python
# Write your code to make your dataset analysis ready.
neighbourhood_mapping = {nb: idx + 1 for idx, nb in enumerate(train['NEIGHBOURHOOD'].dropna().unique())}

# Add a new column for encoded neighborhoods, setting missing values as 0
train['encoded_neighbourhood'] = train['NEIGHBOURHOOD'].map(neighbourhood_mapping).fillna(0).astype(int)

# Fill missing values in 'NEIGHBOURHOOD' and 'HUNDRED_BLOCK' with "unknown"
train[['NEIGHBOURHOOD', 'HUNDRED_BLOCK']] = train[['NEIGHBOURHOOD', 'HUNDRED_BLOCK']].fillna('unknown')
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H): There is no significant association between the neighborhood and crime type.

Alternative Hypothesis (H): There is a significant association between the neighborhood and crime type.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
contingency_table = pd.crosstab(train['NEIGHBOURHOOD'], train['TYPE'])

# Performing Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Significant association between NEIGHBOURHOOD and Crime TYPE (Reject Null Hypothesis)")
else:
    print("No significant association (Fail to Reject Null Hypothesis)")
````

##### Which statistical test have you done to obtain P-Value?

The statistical test used is the Chi-Square Test of Independence.

##### Why did you choose the specific statistical test?

The Chi-Square test is used to determine whether there is a significant association between two categorical variables (Neighborhood and Crime Type).
Since both NEIGHBOURHOOD and TYPE are categorical variables, the Chi-Square test is the appropriate choice.
The p-value obtained is 0.0, which is less than 0.05, leading to the rejection of the null hypothesis. This indicates that there is a significant association between neighborhood and crime type.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H): There is no significant difference in crime occurrence time (HOUR) across different neighborhoods.

Alternative Hypothesis (H): Crime occurrence time (HOUR) significantly varies based on the neighborhood.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
anova_model = ols('HOUR ~ C(NEIGHBOURHOOD)', data=train).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)

print(anova_table)
````

##### Which statistical test have you done to obtain P-Value?

The statistical test used is Analysis of Variance (ANOVA). A one-way ANOVA was conducted to determine whether the mean crime occurrence time differs significantly across various neighborhoods

##### Why did you choose the specific statistical test?

Answer Here.
Since p-value < 0.05, we reject the null hypothesis and conclude that crime occurrence time is significantly influenced by the neighborhood.

### 2. Handling Outliers

````python
num_cols = train.select_dtypes(include=[np.number]).columns

# Compute Q1, Q3, and IQR
Q1 = train[num_cols].quantile(0.25)
Q3 = train[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers using IQR method
outlier_mask = (train[num_cols] < lower_bound) | (train[num_cols] > upper_bound)

# Count outliers per column
outlier_counts = outlier_mask.sum()
outlier_counts = outlier_counts[outlier_counts > 0]


# Handling Outliers

#Clipping: Replace outliers with nearest valid boundary
train[num_cols] = train[num_cols].clip(lower=lower_bound, upper=upper_bound, axis=1)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The outlier treatment techniques used in the given code are:

Interquartile Range (IQR) Method:
The IQR method is used to detect outliers by computing the 25th percentile (Q1) and the 75th percentile (Q3) of the numerical columns.
The lower and upper bounds for outliers are defined as:
Lower Bound = Q1 - 1.5 * IQR
Upper Bound = Q3 + 1.5 * IQR
Any values outside this range are considered outliers.
Clipping:
Instead of removing the outliers, the values are clipped to the nearest valid boundary.
This means any values below the lower bound are replaced with the lower bound, and any values above the upper bound are replaced with the upper bound.
Clipping helps retain all data points while limiting the impact of extreme values.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
import numpy as np

# Prepare Data
crime_counts_daily = train.resample('D', on='Date').size()
X = np.arange(len(crime_counts_daily)).reshape(-1, 1)  # Days as feature
y = crime_counts_daily.values

# Time Series Cross-Validation
tscv = TimeSeriesSplit(n_splits=5)
mse_scores = []

for train_index, test_index in tscv.split(X):
    train, test = y[train_index], y[test_index]

    # Fit ETS Model
    ets_model = ExponentialSmoothing(train, trend="add", seasonal=None, damped_trend=True)
    ets_fit = ets_model.fit()

    # Predict & Evaluate
    predictions = ets_fit.forecast(len(test))
    mse = mean_squared_error(test, predictions)
    mse_scores.append(mse)

# Print Cross-Validation Results
print(f"ETS Cross-Validation MSE Scores: {mse_scores}")
print(f"Average MSE: {np.mean(mse_scores):.2f}")

# Fit the Algorithm
from itertools import product

# Define Parameter Grid
trend_options = ["add", "mul", None]
seasonal_options = ["add", "mul", None]
param_grid = list(product(trend_options, seasonal_options))

best_mse = float("inf")
best_params = None

for trend, seasonal in param_grid:
    try:
        ets_model = ExponentialSmoothing(crime_counts_daily, trend=trend, seasonal=seasonal, damped_trend=True)
        ets_fit = ets_model.fit()
        predictions = ets_fit.forecast(steps=30)
        mse = mean_squared_error(crime_counts_daily[-30:], predictions)
        
        if mse < best_mse:
            best_mse = mse
            best_params = (trend, seasonal)

    except:
        continue

print(f"Best ETS Parameters: Trend = {best_params[0]}, Seasonal = {best_params[1]}")
print(f"Best ETS Model MSE: {best_mse:.2f}")


# Predict on the model
````
